<b> A function is defined, that takes as input the borough as well as the latitude and longitude and gives back the venues around the location provided:<b>

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### a) Johns favourite venues ( mia_venueFAV )
As we already know, that John prefers to have Art Gallery, coffee stores, bars, restaurants and Gyms nearby, we use the dataframe mia_venueFAV that just containes these venues. Using one hot encoding, we will identify how many of the favourite venues are located in the different neigborhoods. A "Favorite score" is defined building the normalized sum of favourite venues in the neigborhood. The following steps are performed:
   1. One hot encoding of Johns favourite venues.
   2. Smaller modifications.
   3. Group the dataframe based on the neigborhood and calculate the mean of the specific venue within this neigborhood

```python
# Step 1: One hot encoding of Johns favourite venues 
mia_venueFAV_onehot = pd.get_dummies(mia_venueFAV[['Venue Category']], prefix="", prefix_sep="")
# Step 2: Smaller modifications 
# add neighborhood column back to dataframe
mia_venueFAV_onehot['Neighborhood'] = mia_venueFAV['Neighborhood']
# move neighborhood column to the first column
fixed_columns = [mia_venueFAV_onehot.columns[-1]] + list(mia_venueFAV_onehot.columns[:-1])
mia_venueFAV_onehot = mia_venueFAV_onehot[fixed_columns]
```

```python
mia_venueFAV_onehot.head()
```

```python
#Step 3: Group the dataframe based on the neigborhood and calculate the sum of the specific venue within this neigborhood
mia_venueFAV_grouped = mia_venueFAV_onehot.groupby('Neighborhood').sum().reset_index()
mia_venueFAV_grouped.head()
```

we need to modify this dataframe by summing up all of the sub-categories to the the main categories

```python
#Step 1: Create a reference dataframe (_ mia_venueFAV_grouped_sum_ ), that contains the same indexing as 
#our dataframe _mia_venueFAV_grouped_
mia_venueFAV_grouped_sum = pd.DataFrame()
mia_venueFAV_grouped_sum = mia_venueFAV_grouped.iloc[:, :1]

#Step 2: Create a separate dataframe for every category containing the sum of the specific category . 
#It is important to keep the indexing from the riginal dataframe
# bars
bars = mia_venueFAV_grouped.filter(like='Bar',axis=1).sum(axis = 1).reset_index(name ='Bars')
bars.set_index(['index'], inplace = True)
# Art Gallery
art = mia_venueFAV_grouped.filter(like='Art Gallery',axis=1).sum(axis = 1).reset_index(name ='Art Gallery')
art.set_index(['index'], inplace = True)
# Cafés
coffees = mia_venueFAV_grouped.filter(like='Café',axis=1).sum(axis = 1).reset_index(name ='Cafés')
coffees.set_index(['index'], inplace = True)
# Gym
gym = mia_venueFAV_grouped.filter(like='Gym',axis=1).sum(axis = 1).reset_index(name ='Gym')
gym.set_index(['index'], inplace = True)
# Restaurants
restaurants = mia_venueFAV_grouped.filter(like='Restaurant',axis=1).sum(axis = 1).reset_index(name ='Restaurants')
restaurants.set_index(['index'], inplace = True)

#Step 3: Combine all of the dataframes based on their index
mia_venueFAV_grouped_sum = pd.concat([mia_venueFAV_grouped_sum, bars, art, coffees, gym, restaurants], axis=1)
mia_venueFAV_grouped_sum.head()
```

```python
#build sum
mia_venueFAV_grouped_sum['FavoriteScore'] = mia_venueFAV_grouped_sum.sum(axis=1)

#normalize values
mia_venueFAV_grouped_sum['FavoriteScore'] = mia_venueFAV_grouped_sum['FavoriteScore']/mia_venueFAV_grouped_sum["FavoriteScore"].sum()

#sort dataframe
mia_venueFAV_grouped_sum.sort_values("FavoriteScore", ascending = False, inplace = True)
mia_venueFAV_grouped_sum.reset_index(drop = True).head(10)
```

```python
#Step 4: Define function that identifies the most common venues in the neigborhoods
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#Step 5: Identify the most common venues in the neighboarhood using the function defned in step 4
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
Common_venueother = pd.DataFrame(columns=columns)
Common_venueother['Neighborhood'] = mia_venueother_grouped['Neighborhood']

for ind in np.arange(mia_venueother_grouped.shape[0]):
    Common_venueother.iloc[ind, 1:] = return_most_common_venues(mia_venueother_grouped.iloc[ind, :], num_top_venues)

Common_venueother.head()

```

## Cluster Neighborhoods
Run k-means to cluster the neighborhood into clusters for all of the venues in the boroughs ( mia_venues ). but first we need to find k means optimal value.

```python
#Step 1: One hot encoding of the venues in munich
mia_venues_onehot = pd.get_dummies(mia_venues[['Venue Category']], prefix="", prefix_sep="")

#Step 2: Smaller modifications 
########add neighborhood column back to dataframe
mia_venues_onehot['Neighborhood'] = mia_venues['Neighborhood'] 

########move neighborhood column to the first column
fixed_columns = [mia_venues_onehot.columns[-1]] + list(mia_venues_onehot.columns[:-1])
mia_venues_onehot = mia_venues_onehot[fixed_columns]
```

```python
#Step 3: Group the dataframe based on the neigborhood and calculate the mean of the specific venue within this neigborhood
mia_venues_grouped = mia_venues_onehot.groupby('Neighborhood').mean().reset_index()
mia_venues_grouped.head()

```

```python
# Drop neigborhood column as it should not be considered in the clustering
mai_venues_grouped_clustering = mia_venues_grouped.drop(['Neighborhood'], 1)
```

```python
# Determine the optimal value of K in K-Means Clustering
cost =[]
for i in range(1, 10):
    KM = KMeans(n_clusters = i, max_iter = 500)
    KM.fit(mai_venues_grouped_clustering)
      
    # calculates squared error
    # for the clustered points
    cost.append(KM.inertia_)     
  
# plot the cost against K values
plt.plot(range(1, 10), cost, color ='g', linewidth ='3')
plt.xlabel("Value of K")
plt.ylabel("Sqaured Error (Cost)")
plt.show() # clear the plot
```

```python
#Step 1: Set number of clusters
kclusters = 3

#Step 2: Run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(mai_venues_grouped_clustering)
```
